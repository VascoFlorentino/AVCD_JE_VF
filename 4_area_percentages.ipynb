# Area proportion of crops in freguesias

## Permanent Crops

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Read the data from the CSV file
data = pd.read_csv('area_perm.csv')

# Get unique crop types and their associated colors
crop_types = data['Temp_crop'].unique()
crop_colors = {
    'Olive plantations': 'green',
    'Fresh fruit plantations (excluding citrus plantations)': 'orange',
    'Vineyards': 'purple',
    'Nuts plantations': 'brown'
}

# Create a figure and axis
fig, ax = plt.subplots(figsize=(12, 6))

# Iterate over the crop types and plot bars for each region
for crop in crop_types:
    crop_data = data[data['Temp_crop'] == crop]
    region_names = crop_data['region_name'] + '_' + crop_data['n'].astype(str)
    perc_area = crop_data['perc_area']
    color = crop_colors[crop]

    ax.bar(region_names, perc_area, color=color, label=crop)

# Set the labels and title
ax.set_xlabel('Region Name')
ax.set_ylabel('Percentage Area')
ax.set_title('Percentage Area (above 1%) of Crops Permanent by Freguesia')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=50, fontsize=8, ha='right', va='top')

# Show the legend
ax.legend()

# Save the last image to the "images" folder
plt.savefig('images/perc_area_perm.png')

# Display the plot
plt.show()
```

## Temporary Crops

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Read the data from the CSV file
data = pd.read_csv('area_temp.csv')

# Define crop colors
crop_colors = {
    'Olive plantations': 'green',
    'Fresh fruit plantations (excluding citrus plantations)': 'orange',
    'Vineyards': 'purple',
    'Nuts plantations': 'brown',
    'Cereals': 'blue',
    'Fodder plants': 'red',
    'Industrial crops': 'yellow'
}

# Create a figure and axis
fig, ax = plt.subplots(figsize=(12, 6))

# Iterate over the crop types and plot bars for each region
for crop in crop_types:
    crop_data = data[data['Temp_crop'] == crop]
    region_names = crop_data['region_name'] + '_' + crop_data['n'].astype(str)
    perc_area = crop_data['perc_area']
    color = crop_colors[crop]

    ax.bar(region_names, perc_area, color=color, label=crop)

# Set the labels and title
ax.set_xlabel('Region Name')
ax.set_ylabel('Percentage Area')
ax.set_title('Percentage Area (above 1%) of Crops Temporary by Freguesia')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=50, fontsize=8, ha='right', va='top')

# Show the legend
ax.legend()

# Save the last image to the "images" folder
plt.savefig('images/perc_area_temp.png')

# Display the plot
plt.show()
```
