### PCA permanent crops

```python
import pandas as pd
import plotly.express as px
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Read the data from CSV file
data = pd.read_csv('3_perm_crops.csv')

# Select the relevant features for PCA
features = ['Area_1989', 'Area_1999', 'Area_2009', 'Area_2019']

# Perform PCA
pca = PCA(n_components=2)
pca_result = pca.fit_transform(data[features])

# Create a DataFrame with PCA results
pca_df = pd.DataFrame(data=pca_result, columns=['PC1', 'PC2'])
pca_df['Perm_crop'] = data['Perm_crop']

# Create a scatter plot with color-coded permanent crops
fig = px.scatter(
    pca_df,
    x='PC1',
    y='PC2',
    color='Perm_crop',
    labels={'Perm_crop': 'Permanent Crop'},
    title='PCA Scatter Plot'
)

fig.show()
```
