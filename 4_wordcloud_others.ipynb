### Create a wordcloud for our report cover:

```python
# Libraries
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Create a list of word
text=("Citrus Olive Fruit Nuts Cereals Dried Pulses Fodder Plants Flowers Ornamental Potatoes Sugarbeets Temporary Grasses Grazings Vineyards")

# Create the wordcloud object
wordcloud = WordCloud(width=480, height=480, margin=0, background_color="white").generate(text)

# Display the generated image
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.margins(x=0, y=0)
plt.show()
```

### Scatterplot Permanent crops

```python
import matplotlib.pyplot as plt
import pandas as pd

# Load the data from a CSV file
data = pd.read_csv('3_perm_crops.csv')

# Convert 'region_name' column to category data type
data['region_name'] = data['region_name'].astype('category')

# Set the figure size
plt.figure(figsize=(10, 10))

# Scatterplot
plt.scatter(
    x=data['Perm_crop'],
    y=data['Area_1989'],
    c=data['region_name'].cat.codes,
    cmap="Accent",
    alpha=0.6,
    edgecolors="white",
    linewidth=2
)

# Set the x-axis and y-axis limits
plt.ylim(0, 12000)  # Adjust the limits as per your data

# Add titles (main and on axis)
plt.xlabel("Perm_crop")
plt.ylabel("Area_1989")
plt.title("Scatterplot")
plt.xticks(rotation=90)

# Display the plot
plt.show()

```

### Violin Plot Permanent crops

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read the data from the CSV file
data = pd.read_csv('3_perm_crops.csv')

# Extract the required column for the violin plot (e.g., Area_1989, Area_1999, etc.)
area_columns = ['Area_1989', 'Area_1999', 'Area_2009', 'Area_2019']
data_to_plot = data[area_columns]

# Melt the data to create a long-form for the violin plot
melted_data = data_to_plot.melt(var_name='Year', value_name='Area')

# Create the violin plot
sns.violinplot(x='Year', y='Area', data=melted_data)

# Customize the plot
plt.xlabel('Year')
plt.ylabel('Area')

# Show the plot
plt.show()
```
