```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data from a CSV file
df = pd.read_csv('3_perm_crops.csv')

# Filter the DataFrame to include only the necessary columns for crop analysis
crop_data = df[['region_name', 'Perm_crop', 'Area_2019']].dropna()

# Pivot the DataFrame to have crop areas as columns
crop_pivot = crop_data.pivot_table(index='region_name', columns='Perm_crop', values='Area_2019', aggfunc='sum')

# Calculate the correlations between crop areas
crop_correlations = crop_pivot.corr()

# Create a correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(crop_correlations, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5, vmin=0, vmax=1)
plt.title('Correlations between Crop Areas')
plt.xlabel('Crops')
plt.ylabel('Crops')
plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.show()
```
